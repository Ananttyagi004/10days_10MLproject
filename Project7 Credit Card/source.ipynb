```python
data.info()
```

```python
#Checking the number of missing values in the dataset.
data.isnull().sum()
```

```python
#Checking the distribution of fraud and legit transactions
data['Class'].value_counts()
```

```python
#This dataset is highly balanced
# 0--> Normal Transaction
# 1--> Fraud Transaction
```

```python
legit=data[data.Class==0]
fraud=data[data.Class==1]
```

```python
legit
```

```python
fraud
```

```python
legit=legit.sample(n=492)
```

```python
legit
```

```python
data_new=pd.concat([legit,fraud],axis=0)
```

```python
data_new.head()
```

```python
data_new['Class'].value_counts()
```

```python
X=data_new.drop(columns=['Class'],axis=1)
```

```python
Y=data_new['Class']
```

```python
#Test train split of data
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rs=RandomForestClassifier(n_estimators=100)
```

```python
rs.fit(x_train,y_train)
```

```python
predict=rs.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,predict))
```
